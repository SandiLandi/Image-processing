```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import skimage.io as io
import skimage.exposure as exp
from skimage.color import rgb2gray
from skimage import img_as_float
from skimage import data
```

Punkt 1 - histogram

```python
raccoon_path = r"C:\Users\HP\Desktop\Studia Inżynieria\R2 s2\Przetwarzanie obrazów i grafika komputerowa\Labolatoria\lab3\dane3\raccoon.png"
szop = io.imread(raccoon_path)
szop_gray = rgb2gray(szop)

values, bins = exp.histogram(szop_gray)

plt.figure()
plt.bar(bins, values)
plt.title('Histogram monochromatycznego obrazku szopa')
plt.xlabel('Poziomy jasności')
plt.ylabel('liczba pikseli')
```

Pytanie 1 

Histogram bimodalny powinien mieć dwa wyraźne szczyty. Histogram który otrzymaliśmy powyżej nie wygląda na bimodalny ponieważ brakuje wyraźnej "doliny". Lewa wyższa część histogramu reprezentuje ciemniejsze elementy obrazu, które moga odpowiadać cieniom lub ciemniejszym elementą futra.

Pytanie 3 

Histogram o większej liczbie przedziałów 253 ma węższe przedziały, więc każdy słupek reprezentuje mniejszą liczbę pikseli.
Aby porównać go z histogramem 10 przedziałowym, należy znormalizować wysokość słupków, co oznacza przemnożenie go przez stosunek liczby przedziałów. Dzięki temu wartości są przeliczone na równoważną reprezentację liczby pikseli, co pozwala na uczciwe porównanie.

Pytanie 6

Histogram obrazu ze słabym kontrastem jest zawężony – wartości jasności są skupione w wąskim zakresie.

```python
gamma2 = exp.adjust_gamma(szop, 2)
gamma5 = exp.adjust_gamma(szop, 5)

plt.figure()

plt.subplot(3, 1, 1)
plt.imshow(szop)
plt.title('obraz orginalny')

plt.subplot(3, 1, 2)
plt.imshow(gamma2)
plt.title('Gamma 2')

plt.subplot(3, 1, 3)
plt.imshow(gamma5)
plt.title('Gamma 5')

plt.tight_layout()

log2 = exp.adjust_log(szop, 2)
log5 = exp.adjust_log(szop, 5)

plt.figure()

plt.subplot(3, 1, 1)
plt.imshow(szop)
plt.title('obraz orginalny')

plt.subplot(3, 1, 2)
plt.imshow(log2)
plt.title('wzmocnienie 2')

plt.subplot(3, 1, 3)
plt.imshow(log5)
plt.title('wzmocnienie 5')

plt.tight_layout()
```
