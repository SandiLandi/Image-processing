```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread('pl.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
r, g, b = cv2.split(img_rgb)



szary = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(r, cmap='Reds')
plt.axis('off')
plt.show()
```

```python
plt.imshow(g, cmap='Greens')
plt.axis('off')
plt.show()
```

```python
plt.imshow(b, cmap='Blues')
plt.axis('off')
plt.show()
```

```python
plt.imshow(szary, cmap='grey')
plt.axis('off')
plt.show()
```
